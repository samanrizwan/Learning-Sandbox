```python
import pandas as pd
```

```python
#Data sets in Pandas are usually multi-dimensional tables, called DataFrames.
#Series is like a column, a DataFrame is the whole table.
dataset = {
    "student" : ["Name","Age","Gender"],
    "Record" : ["saman", "29", "Female"]
 }
 
df = pd.DataFrame(dataset)
print(df)
```

```python
#Pandas use the loc attribute to return one or more specified row(s) refering to the row index:
print(df.loc[0])
print(df.loc[[0, 1]])
```

```python

df = pd.DataFrame(dataset, index = ["Name", "Age", "Gender"])  #you can name your own index using index parameter

print(df) 

print(df.loc["Age"]) #access using index name
```

```python
#A Pandas Series is like a column in a table.
#It is a one-dimensional array holding data of any type

a = [1,2,3,4]
#myvar = pd.series(a) capital s
myvar = pd.Series(a)
print(myvar)

#if nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.
#This label can be used to access a specified value.

print(myvar[0]) #labels used to access values 

mylabvar = pd.Series(a, index = ['a','b','c','d']) #index parameter used to label values
print(mylabvar)
print(mylabvar["b"]) #access variable using label
```

```python
#function to read the actual data into a DataFrame 

df = pd.read_csv('D:\\ML\\data.csv')  #load csv (comma separated file) containing data
print(df)  
#for large dataframe, pandas will load only first five and last five data rows
```

```python
df.shape #to see how large is dataset
```

```python
print(df.to_string())  #print entire dataset
```

```python
print(df.info())
print(df.head()) #print first five records
print(df.tail()) #print last five records
```

```python
print(df.isna())
#isnull() is another option
```

```python
#Empty cells can potentially give you a wrong result when you analyze data.
new_df = df.dropna()  #remove rows that contain empty cells
print(new_df) # the dropna() method returns a new DataFrame, and will not change the original.

#f you want to change the original DataFrame, use the inplace = True argument
#Remove all rows with NULL values
#df.dropna(inplace = True) 
#print(df)
```

```python
#Another way of dealing with empty cells is to insert a new value instead.

#This way you do not have to delete entire rows just because of some empty cells.

#The fillna() method allows us to replace empty cells with a value

df.fillna(1,inplace=True) #replace empty values in entire data frame
df.fillna({"Calories": 130}, inplace=True) #replace values in specific column
```

```python
print(df.duplicated()) #Returns True for every row that is a duplicate, otherwise False
# if duplicate exist, To remove duplicates, use the drop_duplicates() method
#df.drop_duplicates(inplace = True)  
```

```python
df.corr() #calculates the relationship between each column in your data set
```
